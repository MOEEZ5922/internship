# Prepare patient dataframe for Excel analysis
This notebook prepare the dataframe for the patients like the sexe distribution and age.
This notebook shouls be used for One Sleep analysis

## Load data files

```python
import pandas as pd

# %% Define constant
RAW_DATA_PATH = './data'
# load patient data frame
patient_df = pd.read_csv(RAW_DATA_PATH + "/patient.csv", sep=",")
# Load ONe-Sleep patient list
one_sleep_tracking_df = pd.read_csv(RAW_DATA_PATH + "/Tracking SMS OneSleep.csv", sep=",")
# Load Patient KPI
patient_kpi_df = pd.read_csv(RAW_DATA_PATH + "/patientKPI.csv", sep=",")
patient_kpi_df.head()
```

## Select only one patient in the overall patient data frame

```python
patient_df['Pat_Id_int'] = patient_df['Pat_Id_int'].astype(int)
one_sleep_tracking_df['TRC_Ptt'] = one_sleep_tracking_df['TRC_Ptt'].astype(int)
one_sleep_patient_df = patient_df[patient_df['Pat_Id_int'].isin(one_sleep_tracking_df['TRC_Ptt'])]
one_sleep_patient_df
```

## Select only KPIs for the ONe-Sleep patients

```python
patient_kpi_df['Pat_Id_int'] = patient_kpi_df['Pat_Id_int'].astype(int)
patient_kpi_df = patient_kpi_df[patient_kpi_df['Pat_Id_int'].isin(one_sleep_tracking_df['TRC_Ptt'])]
patient_kpi_df.head()
```

## Save dataframes

```python
PROCESS_DATA_PATH = './data/process'
patient_kpi_df.to_csv(PROCESS_DATA_PATH + '/patient_kpi.csv') 
one_sleep_patient_df.to_csv(PROCESS_DATA_PATH + '/one_sleep_patient.csv') 
```
